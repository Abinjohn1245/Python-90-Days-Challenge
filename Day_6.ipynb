# 9. Scope and Global Variables

```python
def test(x=10):
  print(x)
```

```python
test()
```

```python
test(11)
```

```python
# global var = defined outside the function
x=100


def test(y):
  global x
  print(x, y)

  test(10)
```

```python
n=23

def test(n):
  n=21
  print(n)
```

```python
test(18)
```

# 10. Nested Functions

```python
def hello(name):
  print(f"Hello {name}")
```

```python
hello("Abin John")
```

```python
def fav_city(city_name, name):
  hello(name)
  print(f"Welcome to {city_name}!")
```

```python
fav_city("Mumbai", "David")
```
